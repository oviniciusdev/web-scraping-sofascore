```python
import requests
import pandas as pd

url = 'https://www.sofascore.com/api/v1/event/12721041/statistics'

#headers user agent
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36"
}
response = requests.get(url, headers=headers) # Changed request to requests

if response.status_code == 200:
    data = response.json()

    all_list = []

    for match_stats in data['statistics']:
      period = match_stats['period']

      for group in match_stats['groups']:
          group_name = group['groupName'] # Removed extra brackets

          for item in group ['statisticsItems']:
            stat_name = item.get('name')
            home_value = item.get('home')
            away_value = item.get('away')

            all_list.append({
                'Period': period,
                'Group Name': group_name,
                'Stat Name': stat_name,
                'Home Value': home_value,
                'Away Value': away_value
            })

    # criando df a partir dos dados
    df = pd.DataFrame(all_list)

    excel_file = 'sofascore_data.xlsx'
    df.to_excel(excel_file, index=False)

    print(f'Dados exportados para {excel_file}')

df # Now df is accessible
```
